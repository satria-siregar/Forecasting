```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/satria-siregar/Forecasting/main/groceries%20-%20groceries.csv', sep=',')
df.head()
```

```python
df.shape
```

```python
df.drop('Item(s)', inplace=True, axis=1)
```

```python
df = df.replace(np.nan, "0")
df.head()
```

```python
dfnp = np.where(df=='0',0,1)
print(dfnp)
```

```python
df1 = pd.DataFrame(dfnp, columns = ['Item 1','Item 2','Item 3','Item 4','Item 5','Item 6','Item 7','Item 8','Item 9','Item 10','Item 11','Item 12','Item 13','Item 14','Item 15','Item 16','Item 17','Item 18','Item 19','Item 20','Item 21','Item 22','Item 23','Item 24','Item 25','Item 26','Item 27','Item 28','Item 29','Item 30','Item 31','Item 32'])
df1.head()
```

```python
# Build up the frequent items 
frequent_itemsets = apriori(df1, min_support=0.5, use_colnames=True)
frequent_itemsets.head()
```

```python
# Create the rules
rules = association_rules(frequent_itemsets, metric="support", min_threshold=0)
rules
```

```python
rules[ (rules['support'] >= 0.5) &
       (rules['confidence'] >= 0.8) ]
```
